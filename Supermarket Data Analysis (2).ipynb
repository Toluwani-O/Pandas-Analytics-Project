## Step 1 - Loading the dataset
In this step, I will combine the dataset from each branch (3 branches) into one dataset for easy analysis. Also write the syntaxes that will read multiple files from the current working directory and export a CSV file after combining. 

```python
#  Import Libraries
import os
import glob
import pandas as pd
```

```python
#os.chdir() method in Python is used to change the current working directory to specified path.
os.chdir(r"C:\Users\hp\Desktop\Python Projectcsv files")

#List the files in the current directory
os.listdir()
```

```python
# here we want to look for csv files in the current directory
path = r'C:\Users\hp\Desktop\Python Projectcsv files'
files = glob.glob(path + '/*.csv')

# create empty list to store dataframes
list = []

# loop through list of files and read each one into a dataframe and append to list
for f in files:
    temp_df = pd.read_csv(f)
    list.append(temp_df)
    print(f)
```

```python
# concatenate our list of dataframes into one
df = pd.concat(list, axis=0)
```

## Step 3 - Dealing with DateTime Features

```python
# Import Library
import datetime as dt
```

```python
# convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Check the datatype to confirm if it's in datetime
df.info()
```

```python
# convert the Time column to datetime
df['Time'] = pd.to_datetime(df['Time'])
```

```python
# Check the datatype to confirm if it's in datetime
df.info()
```

### Extract Features from date & time

```python
# Extracting the Year feature from the Date column, and saving to a new Year column
df['Year'] = df['Date'].dt.year
```

```python
# Extracting the Month feature from the Date column, and saving to a new Month column
df['Month'] = df['Date'].dt.month
```

```python
# Extracting the Day feature from the Date column, and saving to a new Day column
df['Day'] = df['Date'].dt.day
```

```python
# Extracting the Hour feature from the Time column and saving to a new Hour column
df['Hour'] = df['Time'].dt.hour
```
